```python
olympics.query('type=="Winter" and medal=="Bronze" and born_country=="POL" and year==1972')
```

```python
olympics[olympics['noc'] == 'URU'].dropna(subset=['medal']).groupby(['born_region'])['medal'].value_counts().unstack(fill_value=0)[['Gold', 'Silver', 'Bronze']].sort_values(['Gold', 'Silver', 'Bronze'], ascending=False)
```

```python
olympics.query('year>=1980 and event.str.contains(", Men") and medal=="Gold"').groupby('discipline')[['height_cm', 'weight_kg']].min().sort_values('height_cm', ascending=False).dropna().tail(20)
```
